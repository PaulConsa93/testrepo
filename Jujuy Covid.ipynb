```python
# Traducir SI/NO a 1/0

df_jujuy =df_jujuy.replace({'fallecido': {'SI': 1, 'NO': 0}, 
                            'cuidado_intensivo': {'SI': 1, 'NO': 0},
                            'asistencia_respiratoria_mecanica': {'SI': 1, 'NO': 0}})
```

```python
# Reemplazar Null/NaN/NaT en fecha_diagnostico, con fecha_apertura 

df_jujuy["fecha_diagnostico"] = df_jujuy["fecha_diagnostico"].fillna(df_jujuy["fecha_apertura"])
```

```python
# Dar formato de "Datetime"

df_jujuy["fecha_diagnostico"] = pd.to_datetime(df_jujuy["fecha_diagnostico"])
df_jujuy["fecha_apertura"] = pd.to_datetime(df_jujuy["fecha_apertura"])
df_jujuy["fecha_inicio_sintomas"] = pd.to_datetime(df_jujuy["fecha_inicio_sintomas"])

df_jujuy = df_jujuy.reset_index(drop = True).set_index("fecha_diagnostico").sort_index()
```

```python
# Confirmados en Jujuy
# Se tomará la edad de los niños menor a 1 año como el promedio de la edad en años

bebes_jujuy = df_jujuy[(df_jujuy.clasificacion_resumen == "Confirmado") & (df_jujuy.edad_años_meses == "Meses")]
bebes_edad_promedio = bebes_jujuy["edad"].mean()/12

confirmados_jujuy = df_jujuy[df_jujuy["clasificacion_resumen"] == "Confirmado"]
confirmados_jujuy.loc[confirmados_jujuy["edad_años_meses"] == "Meses", "edad"] = bebes_edad_promedio
confirmados_jujuy = confirmados_jujuy.replace({'edad_años_meses': {'Meses': "años"}})
```
